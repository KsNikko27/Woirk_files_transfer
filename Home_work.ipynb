```python
import numpy as np
import pandas as pd
from numpy.linalg import inv
import matplotlib.pyplot as plt
```

```python
x_hw = np.array([[1,50],
                [1,60],
                [1,70],
                [1,100]])
```

```python
y_hw = np.array([[10],
                [15],
                [40],
                [45]])
```

```python
plt.scatter(x_hw[:,1],y_hw)
```

```python
w = inv(x_hw.T.dot(x_hw)).dot(x_hw.T).dot(y_hw)
```

```python
w
```

```python
x_min = x_hw[:,1].min() -10
x_max = x_hw[:,1].max() + 10
```

```python
X = np.linspace(x_min, x_max,100)
len(X)
```

```python
Y = w[0][0] + w[1][0] * X
```

```python
plt.xlim(x_min, x_max)
plt.ylim(0, y_hw[:].max())
plt.plot(X, Y)
plt.scatter(x_hw[:,1],y_hw)
```
