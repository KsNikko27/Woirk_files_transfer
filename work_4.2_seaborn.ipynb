```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
reviews = pd.read_csv("winemag-data_first150k.csv.gz", compression="gzip")
reviews.head()
```

```python
fig, axes = plt.subplots(2,1, figsize=(18,8))
reviews["points"].value_counts().sort_index().plot.bar(ax=axes[0])
reviews['province'].value_counts().head(20).plot.bar(ax=axes[1])
```

```python
fig, axes = plt.subplots(2, 2, figsize=(18,9))
reviews['points'].value_counts().sort_index().plot.bar(ax=axes[0][0])
reviews['province'].value_counts().head(20).plot.bar(ax=axes[1][1])
```

```python
fig, axes = plt.subplots(2, 2, figsize=(18,9))
reviews['points'].value_counts().sort_index().plot.bar(ax=axes[0][0], fontsize=12, color='forestgreen')
axes[0][0].set_title("Wine Scores", fontsize=18)
reviews['variety'].value_counts().head(20).plot.bar(ax=axes[1][0], fontsize=12, color='olive')
axes[1][0].set_title("Wine Variety", fontsize=18)
reviews['province'].value_counts().head(20).plot.bar(ax=axes[1][1], fontsize=12, color='olive')
axes[1][1].set_title("Wine Province", fontsize=18)
reviews['price'].value_counts().head(20).plot.bar(ax=axes[0][1], fontsize=12, color='forestgreen')
axes[0][1].set_title("Wine Price", fontsize=18)
plt.subplots_adjust(hspace=.3)

sns.despine()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pylab import rcParams
rcParams['figure.figsize'] = 6, 4
```

```python
reviews = pd.read_csv("winemag-data_first150k.csv.gz", compression="gzip")
reviews.head()
```

```python
sns.countplot(x = 'points', data=reviews)
```

```python
sns.kdeplot(reviews.query("price < 100 ").price)
```

```python
sns.distplot(reviews[(-reviews['price'].isna())&(reviews["price"]<100)]['price'], kde=1, bins=10)
```

```python
%config InlineBackend.figure_format = "png"
```

```python
sns.jointplot(data=reviews[reviews['price']<100].sample(500), x='price', y='points')
```

```python
sns.jointplot(data=reviews[reviews['price']<100], x='price', y='points', kind="hex",gridsize=20)
```

```python
sns.kdeplot(x='price', y='points', data=reviews[reviews['price']<100].loc[:,['price','points']].dropna().sample(5000),fill=True)
```

```python
reviews.head()
```

```python
top_variety = reviews[reviews.variety.isin(reviews.variety.value_counts().head(5).index)]
top_variety.points.describe()
```

```python
fig, axes = plt.subplots(figsize=(10,10))
sns.boxplot(x='variety',
            y='points',
           data=top_variety)
```

```python
fig, axes = plt.subplots(figsize=(12,6))
sns.violinplot(data=top_variety,
              x="variety",
              y="points")
```
